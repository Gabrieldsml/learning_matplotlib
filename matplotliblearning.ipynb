```python
# OO method

```

```python
fig, ax = plt.subplots(figsize=(10,6))
#over_50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target', ax = ax)
# ax.set_xlim([45,100]) if you want to create X limits
```

```python
fig, ax = plt.subplots(figsize = (10,6))
scatter = ax.scatter(x=over_50["age"], y = over_50["chol"], c = over_50["target"], cmap = "plasma")
ax.set(title = "Heaart Disease and Cholesterol Levels", xlabel = "Age", ylabel = "Cholesterol")
ax.set_ylim(100, 600)
ax.axhline(over_50["chol"].mean(), linestyle = "--")
ax.legend(*scatter.legend_elements(), title = "Target")
```

```python
over_50.target.values
```

```python
over_50.head()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10))
t1 = ax1.scatter(over_50["age"], over_50["chol"], c = over_50.target, cmap = "plasma")
ax1.set(title = "Heart Disease and Cholesterol Levels", ylabel = "Cholesterol")
ax1.set_ylim(100, 600)
ax1.set_xlim(50,80)
ax1.axhline(over_50["chol"].mean(), linestyle = "--")
#ax1.legend(*scatter.legend_elements(), title = "teste")
fig.suptitle("Heart Disease Analysis", fontsize = 16)
fig.legend(*t1.legend_elements(), loc=('upper right'), title = "Target")




ax2.set(title = "Heart Disease and Max Heart Rate levels", xlabel = "Age", ylabel = "MH Rate")
scatter = ax2.scatter(x= over_50.age, y = over_50.thalach, c  = over_50.target, cmap = "plasma")
ax2.axhline(over_50.thalach.mean(), linestyle = "--")
ax2.set_ylim(50, 200)
ax2.set_xlim(50,80);

fig.savefig("heart-disease-analysis.png")


```
